```python

import pandas as pd

def get_data(targetdir):
    #list the files
    filelist = os.listdir(targetdir) 
    #read them into pandas
    df_list = [pd.read_csv(targetdir + file) for file in filelist]
    #concatenate them together
    big_df = pd.concat(df_list)
    return big_df
```
